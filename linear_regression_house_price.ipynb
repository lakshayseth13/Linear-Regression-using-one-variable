```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("/data.txt", delimiter = ",")
```

```python
data
```

```python
X = data[:,0]
Y = data[:,1].reshape(X.size,1)
X = np.vstack((np.ones((X.size, )), X)).T
```

```python
plt.scatter(X[:, 1], Y)
plt.show()
```

```python

def model(X , Y, lrate, iteration):
  m =Y.size
  theta = np.zeros((2,1))
  costlist = []
  theta_history = []
  theta_history.append(theta)
  for i in range(iteration):
    ypred = np.dot(X, theta)
    cost = (1/(2*m))*np.sum(np.square(ypred - Y))
    dtheta = (1/m)*np.dot(X.T, ypred - Y)
    theta = theta - lrate*dtheta
    theta_history.append(theta)
    costlist.append(cost)
  return theta, costlist, theta_history
```

```python
iteration = 100
lrate = 0.00000005
theta, cost_list, thetahistory = model(X, Y, lrate, iteration)
ypred = np.dot(X, theta)
plt.scatter(X[:, 1], Y)
plt.plot(X[:,1], ypred)
plt.show()
```
